# Selective Fit

```python
import numpy as np
from smpl import plot
from smpl import functions as f
import uncertainties.unumpy as unp
```

```python

data = np.loadtxt('test_data.txt')
xdata = data[:,0]
xerr = data[:,2]
ydata = data[:,1]
yerr = data[:,3]
x = unp.uarray(xdata,xerr)
y = unp.uarray(ydata,yerr)
```

```python


ff = plot.fit(xdata, ydata, fmt='.', label='data', xaxis="x in a.u.",yaxis="y in a.u.",function=f.line, frange=[0,6])
ff = plot.fit(xdata, ydata, fmt='.', label='data', xaxis="x in a.u.",yaxis="y in a.u.",function=f.line, frange=[6,10],also_data=False,init=False)
plot.show()
```

```python

ff = plot.fit(x, y, fmt='.', label='data', xaxis="x in a.u.",yaxis="y in a.u.",function=f.line, selector=lambda x,y : x<10)
ff = plot.fit(x, y, fmt='.', label='data', xaxis="x in a.u.",yaxis="y in a.u.",function=f.line, selector=x >10,also_data=False,init=False)
plot.show()
```
