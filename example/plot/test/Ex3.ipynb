```python
from smpl import plot
import numpy as np
```

# $V = -\int \dot x dx$

## $V(x)=-\frac{1}{2}x^2+ \frac{1}{3}x^3$

metastable at $x=0$ for $x\geq 0$

```python

plot.function( lambda x : -1/2 *x**2+1/3*x**3, xaxis="$x$", yaxis="$V(x)$",xmin=-10, xmax=10 )
```

## $V(x) = -3x$

No fixed point

```python

plot.function( lambda x : -3*x, xaxis="$x$", yaxis="$V(x)$",xmin=-10, xmax=10 )
```

## $V(x) = cosh(x)$

Stable fixed point at $x=0$

```python

plot.function( lambda x : np.cosh(x), xaxis="$x$", yaxis="$V(x)$",xmin=-2, xmax=2 )
```
