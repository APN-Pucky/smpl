```python
import numpy as np
from smpl import plot
```

```python
def gompertz(N,a,b):
    return -a * N * np.log(b*N)

def solution_gompertz(t,a,b,c):
    return 1/b*np.exp(np.exp(-t*a)*c)
```

# 1 Solution and Plots
Separation of Variables:

$ \int \frac{1}{N \ln (bN)} dN= \ln( \ln( bN)) +c= -adt =-at $

$ \implies N(t) =  \frac{1}{b} \exp (c \exp (-t a)) $

```python
for a in [1,2]:
    for b in [1,0.5]:
        plot.function( gompertz, a,b, xaxis="$N$", yaxis="$\\dot N$",xmin=0.1, xmax=3 ,init=False )
plot.show()

for a in [1,2]:
    for b in [1,1/8]:
        for c in [1,2]:
            plot.function( solution_gompertz, a,b,c, xaxis="$t$", yaxis="$N$",xmin=0.1, xmax=3,init=False )
plot.show()
```

# 2 Fixed points
$ {\ddot N} = -a ( \ln (bN)+1) $

$N = 0$
unstable since $ {\ddot N}(0)  > 0$ given $ N\geq0 $

$N = \frac{1}{b}$
stable since $ {\ddot N}(\frac{1}{b}) = -2 a < 0$

# 3 Bio meaning
$b$ is the inverse of the stable number of cancer tumors for $t \to \infty$.

$a$ defines how fast the cancer grows.
