# Lambda Fit

```python
import numpy as np
from smpl import plot
from smpl import stat
from smpl import functions as f
import uncertainties.unumpy as unp
import smpl
```

String fit does not work with sigma=N, as it needs unp.exp

```python
x= np.linspace(-5,5,10)
y = stat.noisy(np.exp(2*x))+stat.noisy(x)
ff = plot.fit(x, y,"exp(a*x)*c" , fmt='.', label='data', xaxis="x in a.u.",yaxis="y in a.u.")
```

```python
x= np.linspace(-5,5,10)
y = stat.noisy(np.exp(2*x))+stat.noisy(x)
ff = plot.fit(x, y,lambda x,a,c: np.exp(x*a)*c , fmt='.', label='data', xaxis="x in a.u.",yaxis="y in a.u.")
```

```python
x= np.linspace(-5,5,10)
y = stat.noisy(np.exp(2*x))+stat.noisy(x)
ff = plot.fit(x, y,lambda x,a,c: unp.exp(x*a)*c , fmt='.', label='data', xaxis="x in a.u.",yaxis="y in a.u.",sigmas=1)
```
