```python
import smpl
from smpl import plot
from smpl import stat
from smpl import functions as f
import numpy as np
print(smpl.__version__)
```

```python
x= np.random.randn(100000)
# Default uncertainty of bins is poisson distributed in y direction and none for x
plot.fit(x,0,f.gauss,bins=20,label="data",binunc=stat.no_dist,init=True)
plot.fit(x,0,f.gauss,bins=20,label="data",init=True)
plot.fit(x,0,f.gauss,bins=20,label="data",fmt="hist",init=True)
plot.fit(x,0,f.gauss,bins=20,label="data",fmt="step",init=True)
```
